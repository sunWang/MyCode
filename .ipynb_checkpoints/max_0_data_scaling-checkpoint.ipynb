```python
import pandas as pd
import numpy as np
# import seaborn as sns
# import matplotlib.pyplot as plt
import os
from sklearn.model_selection import train_test_split

from sklearn.metrics import roc_curve,  roc_auc_score
# plt.style.use('seaborn-colorblind')
# %matplotlib inline
#from feature_cleaning import rare_values as ra
```

```python
data = pd.read_csv('max_data/SixAirlinesDataV2.csv')
data = data.drop(columns='Unnamed: 0')
data
```

```python
# add the new created feature
from sklearn.preprocessing import StandardScaler
ss = StandardScaler().fit(data[['PriceEconomy']])
data = data.copy(deep=True)
data['PriceEconomy'] = ss.transform(data[['PriceEconomy']])
print(data.head(6))
```

```python
# add the new created feature
from sklearn.preprocessing import StandardScaler

ss = StandardScaler().fit(data[['PricePremium']])
data = data.copy(deep=True)
data['PricePremium'] = ss.transform(data[['PricePremium']])

ss = StandardScaler().fit(data[['SeatsTotal']])
data = data.copy(deep=True)
data['SeatsTotal'] = ss.transform(data[['SeatsTotal']])

ss = StandardScaler().fit(data[['PitchDifference']])
data = data.copy(deep=True)
data['PitchDifference'] = ss.transform(data[['PitchDifference']])

ss = StandardScaler().fit(data[['WidthDifference']])
data = data.copy(deep=True)
data['WidthDifference'] = ss.transform(data[['WidthDifference']])

ss = StandardScaler().fit(data[['month_num']])
data = data.copy(deep=True)
data['month_num'] = ss.transform(data[['month_num']])


ss = StandardScaler().fit(data[['PercentPremiumSeats']])
data = data.copy(deep=True)
data['PercentPremiumSeats'] = ss.transform(data[['PercentPremiumSeats']])

ss = StandardScaler().fit(data[['WidthPremium']])
data = data.copy(deep=True)
data['WidthPremium'] = ss.transform(data[['WidthPremium']])

ss = StandardScaler().fit(data[['WidthEconomy']])
data = data.copy(deep=True)
data['WidthEconomy'] = ss.transform(data[['WidthEconomy']])

ss = StandardScaler().fit(data[['PitchPremium']])
data = data.copy(deep=True)
data['PitchPremium'] = ss.transform(data[['PitchPremium']])

ss = StandardScaler().fit(data[['PitchEconomy']])
data = data.copy(deep=True)
data['PitchEconomy'] = ss.transform(data[['PitchEconomy']])

ss = StandardScaler().fit(data[['SeatsPremium']])
data = data.copy(deep=True)
data['SeatsPremium'] = ss.transform(data[['SeatsPremium']])

ss = StandardScaler().fit(data[['SeatsEconomy']])
data = data.copy(deep=True)
data['SeatsEconomy'] = ss.transform(data[['SeatsEconomy']])

ss = StandardScaler().fit(data[['FlightDuration']])
data = data.copy(deep=True)
data['FlightDuration'] = ss.transform(data[['FlightDuration']])

ss = StandardScaler().fit(data[['Airline']])
data = data.copy(deep=True)
data['Airline'] = ss.transform(data[['Airline']])

ss = StandardScaler().fit(data[['Aircraft']])
data = data.copy(deep=True)
data['Aircraft'] = ss.transform(data[['Aircraft']])

ss = StandardScaler().fit(data[['IsInternational']])
data = data.copy(deep=True)
data['IsInternational'] = ss.transform(data[['IsInternational']])

print(data.head(6))
```
